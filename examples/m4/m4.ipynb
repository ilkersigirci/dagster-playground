# Sktime Training

## Specific time serie

```python
y_train["D1"]
```

```python
unique_id = "D1"

y_train_one = y_train[unique_id]
y_test_one = y_test[unique_id]

train_index = pd.period_range(start="2000-01-01", periods=len(y_train_one), freq="D", name="Period")
test_index = pd.period_range(start=train_index[-1]+1, periods=len(y_test_one), freq="D", name="Period")

y_train_one = pd.Series(y_train_one, index=train_index, name=unique_id)
y_test_one = pd.Series(y_test_one, index=test_index, name=unique_id)
```

```python
y_train_one
```

```python
y_test_one
```

```python
from sktime.forecasting.ets import AutoETS


fh = np.arange(1, len(y_test_one)+1)
sp = info.seasonality
model = AutoETS(sp=sp)

model.fit(y_train_one).predict(fh)
```
